```python
import pandas as pd
```

```python
df = pd.read_csv("1_Petrol_price_districtwise.csv")
df.head()
```

```python
df.dtypes
```

```python
df = df[['District','State','fuel_pumps','modal_price']]
df.head()
```

```python
# maximum price in each state:-

states_max = df.groupby('State').modal_price.max()
states_max
```

```python
# minimum prices in each state:-

states_min = df.groupby('State').modal_price.min()
states_min
```

```python
# Combining both data in a single data frame

states_price = pd.merge(states_min, states_max, on="State")
states_price
```

```python
# Renaming the columns:-

states_price.rename(columns = {'State':'State', 'modal_price_x':'min_price',
                               'modal_price_y':'max_price'}, inplace = True)

states_price
   
```

```python
#copying this to excel:-
states_price.to_clipboard("states_price.xlsx")
```

```python
# Arranging the states/union territories based on max fuel price from high to low 

states_max = states_max.sort_values(ascending = False)
states_max
```
